```python
# dependencies
import pandas as pd
```

```python
# import csv
raw_data = pd.read_csv("Red_Light_Camera_Violations.csv")
raw_data.head()
```

```python
# filter for blanks, nulls
data_clean_columns = raw_data[["CAMERA ID", "VIOLATION DATE", "VIOLATIONS", "LATITUDE", "LONGITUDE", "Zip Codes"]].dropna()
data_clean_columns
```

```python
data_clean_columns2 = data_clean_columns.rename(columns={'VIOLATION DATE': 'VIOLATION_DATE',
                                                        'CAMERA ID': 'CAMERA_ID',
                                                        'Zip Codes': 'Zip_Codes'})
data_clean_columns2
```

```python
# group by: camera id and month / year, etc.
data_clean_columns2['VIOLATION_DATE'] = pd.to_datetime(data_clean_columns2['VIOLATION_DATE'])
data_clean_columns2
```

```python
data_clean_columns2['Year'] = data_clean_columns2['VIOLATION_DATE'].dt.year
data_clean_columns2['Month'] = data_clean_columns2['VIOLATION_DATE'].dt.month
data_clean_columns2
```

```python
data_clean_columns2.to_csv('hopeful.csv', index=False)
```

```python
import csv
import json
```

```python

def csv_to_json(csv_file, json_file):
    # Read CSV file and convert it to a list of dictionaries
    csv_data = []
    with open(csv_file, 'r') as csvfile:
        csv_reader = csv.DictReader(csvfile)
        for row in csv_reader:
            csv_data.append(row)

    # Write the list of dictionaries to a JSON file
    with open(json_file, 'w') as jsonfile:
        jsonfile.write(json.dumps(csv_data, indent=2))

if __name__ == "__main__":
    input_csv_file = "static/hopeful.csv"  # Replace with the path to your CSV file
    output_json_file = "hopeful.json"  # Replace with the desired output JSON file path

    csv_to_json(input_csv_file, output_json_file)
```
